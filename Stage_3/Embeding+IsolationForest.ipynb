```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sentence_transformers import SentenceTransformer

df = pd.read_csv("Unified_Text.csv")
df.head(2)
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
X = model.encode(df['log_text'].tolist(), show_progress_bar=True)
iso = IsolationForest(n_estimators=200, contamination=0.10, random_state=42).fit(X)
df['label'] = iso.predict(X)

df['label'] = df['label'].map({1: 'non_critical', -1: 'anomaly'})


print(df['label'].value_counts())


df.to_csv("Final_Labeled.csv", index=False)
print("Final_Labeled.csv saved successfully!")
```
