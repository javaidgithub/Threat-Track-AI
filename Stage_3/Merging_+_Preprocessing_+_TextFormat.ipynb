```python
import pandas as pd
from pathlib import Path

def parse_access_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'Access'
    df['source_ip'] = df['ip']
    df['component'] = None
    df['level'] = None
    df['status'] = df['status']
    df['event_id'] = None
    df['content'] = df['request']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_apache_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'Apache'
    df['source_ip'] = None
    df['component'] = None
    df['level'] = df['Level']
    df['status'] = None
    df['event_id'] = df['EventId']
    df['content'] = df['Content']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_audit_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'Audit'
    df['source_ip'] = None
    df['component'] = df['controller']
    df['level'] = None
    df['status'] = None
    df['event_id'] = None
    df['content'] = df['action'] + ' - ' + df['actionlog']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_hadoop_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'Hadoop'
    df['source_ip'] = None
    df['component'] = df['Component']
    df['level'] = df['Level']
    df['status'] = None
    df['event_id'] = df['EventId']
    df['content'] = df['Content']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_hdfs_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'HDFS'
    df['source_ip'] = None
    df['component'] = df['Component']
    df['level'] = df['Level']
    df['status'] = None
    df['event_id'] = df['EventId']
    df['content'] = df['Content']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_openstack_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'OpenStack'
    df['source_ip'] = df['ADDR'].str.extract(r'(\d+\.\d+\.\d+\.\d+)')
    df['component'] = df['Component']
    df['level'] = df['Level']
    df['status'] = df['Content'].str.extract(r'status:\s?(\d+)')
    df['event_id'] = df['EventId']
    df['content'] = df['Content']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_web_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'Web'
    df['source_ip'] = df['IP']
    df['component'] = None
    df['level'] = None
    df['status'] = df['Staus']
    df['event_id'] = None
    df['content'] = df['URL']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

def parse_windows_log(file_path):
    df = pd.read_csv(file_path)
    df['log_type'] = 'Windows'
    df['source_ip'] = None
    df['component'] = df['Component']
    df['level'] = df['Level']
    df['status'] = None
    df['event_id'] = df['EventId']
    df['content'] = df['Content']
    return df[['log_type', 'source_ip', 'component', 'level', 'status',
               'event_id', 'content', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

# Load and unify all logs
access_df = parse_access_log("Access_Log.csv")
apache_df = parse_apache_log("Apache_Log.csv")
audit_df = parse_audit_log("Audit_Log.csv")
hadoop_df = parse_hadoop_log("Hadoop_Log.csv")
hdfs_df = parse_hdfs_log("HDFS_Log.csv")
openstack_df = parse_openstack_log("OpenStack_Log.csv")
web_df = parse_web_log("Web_Log.csv")
windows_df = parse_windows_log("Window_Log.csv")

# Combine all into one DataFrame
unified_df = pd.concat([
    access_df, apache_df, audit_df,
    hadoop_df, hdfs_df, openstack_df,
    web_df, windows_df
], ignore_index=True)

# Save unified output
unified_df.to_csv("Unified_Log_Format.csv", index=False)
print("Unified log saved to 'Unified_Log_Format.csv'")
```

```python
import pandas as pd

df = pd.read_csv('Unified_Log_Format.csv')

# Preprocess each row to convert into a text format
def preprocess_log(row):
    def safe_int(val):
        return int(val) if pd.notna(val) else 0

    year = safe_int(row.get('Year'))
    month = safe_int(row.get('Month'))
    day = safe_int(row.get('Day'))
    hour = safe_int(row.get('Hour'))
    minute = safe_int(row.get('Minute'))

    return (
        f"Log Type: {row.get('log_type', 'None')}, "
        f"Source IP: {row.get('source_ip', 'None')}, "
        f"Level: {row.get('level') if pd.notna(row.get('level')) else 'None'}, "
        f"Status: {row.get('status') if pd.notna(row.get('status')) else 'None'}, "
        f"Event ID: {row.get('event_id') if pd.notna(row.get('event_id')) else 'None'}, "
        f"Content: {row.get('content', 'None')}, "
        f"Timestamp: {year}-{month:02d}-{day:02d} {hour:02d}:{minute:02d}"
    )

# Apply preprocessing to all rows
df['log_text'] = df.apply(preprocess_log, axis=1)
```
